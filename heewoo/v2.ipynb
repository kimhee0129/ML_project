```python
# 모듈 임포트
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import cv2, os
from tqdm import tqdm
import numpy as np
from sklearn.model_selection import KFold
```

```python
# 상수 정의
dataset_path = "E:\\AI_data\\training_set_gen"

str2label = { "bird" : 0,
              "car" : 1,
              "cat" : 2,
              "dog" : 3,
              "fish" : 4 }
```

```python
from torchvision.models import resnet18, ResNet18_Weights
from torch import nn, optim
```

```python
# 신경망 정의
class Res18Net(nn.Module):
    def __init__(self):
        super(Res18Net, self).__init__()
        self.model = resnet18(weights=ResNet18_Weights.DEFAULT)
        self.model.conv1 = nn.Conv2d(3, 64, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False)
        num_ftrs = self.model.fc.in_features
        self.model.fc = nn.Linear(num_ftrs, 5)

    def forward(self, x):
        x = self.model(x)
        return x
```

```python
import torch, random
import torchvision.transforms as transforms
import torchvision.datasets as datasets
from torch.utils.data import DataLoader, Subset
from sklearn.metrics import f1_score

num_epochs = 50
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

transform = transforms.Compose([
    transforms.ToTensor(),  # 이미지를 텐서로 변환
])

dataset = datasets.ImageFolder(root=dataset_path, transform=transform)

# 각 클래스의 인덱스 딕셔너리 생성
class_indices = {}
for idx, (_, label) in enumerate(tqdm(dataset)):
    if label not in class_indices:
        class_indices[label] = []
    class_indices[label].append(idx)

# 클래스 균일하게 분할할 비율 설정
train_ratio = 0.8  # 80%를 훈련 데이터로 사용
val_ratio = 0.2    # 20%를 검증 데이터로 사용

train_indices = []
val_indices = []

# 각 클래스에서 인덱스 분할
for indices in class_indices.values():
    random.shuffle(indices)  # 인덱스 무작위로 섞기
    split_point = int(len(indices) * train_ratio)
    train_indices.extend(indices[:split_point])
    val_indices.extend(indices[split_point:])

# 서브셋 생성
train_subset = Subset(dataset, train_indices)
val_subset = Subset(dataset, val_indices)

batch_size = 32
train_loader = DataLoader(train_subset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_subset, batch_size=batch_size, shuffle=False)

```

```python
from sklearn.metrics import f1_score
import time

model = Res18Net().to(device)
criterion = nn.BCEWithLogitsLoss().to(device)
optimizer = optim.Adam(model.parameters(), lr=0.001)

val_loss = []
val_acc = []
val_f1 = []
train_loss = []

num_epochs = 3

patience = 10
max_acc = None
no_progress = 0

for epoch in range(num_epochs):
    start_time = time.time()
    model.train()

    total_train_loss = 0
    for inputs, labels in train_loader:
        one_hot = []
        for l in labels:
            result = [0 for i in range(5)]
            result[l] = 1
            one_hot.append(result)
        one_hot = np.array(one_hot)
        targets = torch.tensor(one_hot, dtype=torch.float32)

        inputs = inputs.to(device)
        targets = targets.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, targets)
        total_train_loss += loss.item()
        loss.backward()
        optimizer.step()

    train_loss.append(total_train_loss / len(train_loader))

    model.eval()
    total_val_loss = 0
    all_preds = []
    all_targets = []
    
    with torch.no_grad():
        for inputs, labels in val_loader:
            one_hot = []
            for l in labels:
                result = [0 for i in range(5)]
                result[l] = 1
                one_hot.append(result)
            one_hot = np.array(one_hot)
            targets = torch.tensor(one_hot, dtype=torch.float32)
            inputs = inputs.to(device)
            targets = targets.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, targets)
            total_val_loss += loss.item()
        
            preds = torch.sigmoid(outputs) > 0.5
            all_preds.extend(preds.cpu().numpy())
            all_targets.extend(targets.cpu().numpy())

    val_loss.append(total_val_loss / len(val_loader))
    
    all_preds = np.array(all_preds)
    all_targets = np.array(all_targets)

    accuracy = (all_preds == all_targets).mean()
    val_acc.append(accuracy)

    f1 = f1_score(all_targets, all_preds, average='weighted')
    val_f1.append(f1)

    ## 저장 and early stopping ##
    if not max_acc:
        max_acc = accuracy

    if max_acc < accuracy:
        no_progress = 0
        max_acc = accuracy
        torch.save(model.state_dict(), f"v2_{epoch}.wgt")
        print(f"model saved! (val_acc: {max_acc})")
    else:
        no_progress += 1

    if no_progress > patience:
        print(f"no progress during {patience} epochs!")
        break

    print(f"Epoch ({epoch + 1} / {num_epochs}) Done in {time.time() - start_time : d} secs.\n[ train_loss = {train_loss[-1]}, val_loss = {val_loss[-1]} ]")
```

```python
def smooth_curve(points, factor=0.9):
    smoothed_points = []
    for point in points:
        if smoothed_points:
            previous = smoothed_points[-1]
            smoothed_points.append(previous * factor + point * (1 - factor))
        else:
            smoothed_points.append(point)
    return smoothed_points
```

```python
import matplotlib.pyplot as plt

plt.plot(range(1, num_epochs+1), train_loss, label='Training loss')
plt.plot(range(1, num_epochs+1), val_loss, label='Validation loss')

plt.xlabel("Epochs")
plt.legend()
plt.show()
```

```python
plt.cla()

plt.plot(range(1, len(val_acc)+1), val_acc, label='Validation accuracy')
plt.plot(range(1, len(val_f1)+1), val_f1, label='Validation f1 score')

plt.xlabel("Epochs")
plt.legend()
plt.show()
```
